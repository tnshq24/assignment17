```python
#Q1
# Matplotlib is used to plot the information which is form of numbers
# at first but after plotting them they can be shown as graphs.

# Some plots are
# Scatter
# Pie
# Bar 
# Area
# Histogram
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
```

```python
#Q2
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
#Q3
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x1,y1)

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x2,y2)

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x3,y3)
```

```python
# Subplot is used to plot muliple graphs at the same time without
# overlapping on one another.
```

```python
#Q4

# Bar plot is a function of matplotlib library which is used 
# to plot the data into bar graph representation.

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
#Q5
# Box is used to represent data in grouped form
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.plot(box1,box2)
plt.box(on=True)
```
